```python
import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.evaluation import evaluate_policy

# Define the source and target environments
source_env_name = "CartPole-v1"  # Source environment
target_env_name = "CartPole-v0"  # Target environment (same observation and action spaces)
```

```python
# Create and train on the source environment (CartPole-v1)
source_env = gym.make(source_env_name)
model = PPO("MlpPolicy", source_env, verbose=1)
model.learn(total_timesteps=20000)  # Pre-train on source task (e.g., 20,000 steps)

# Evaluate the source-trained model
mean_reward, std_reward = evaluate_policy(model, source_env, n_eval_episodes=10)
print(f"Mean reward in source environment ({source_env_name}): {mean_reward} +/- {std_reward}")

# Save the model
model.save("ppo_cartpole")
```

```python
# Load the saved model and transfer to target environment (CartPole-v0)
target_env = gym.make(target_env_name)
model = PPO.load("ppo_cartpole", env=target_env)

# Continue training on the target environment
model.learn(total_timesteps=20000)  # Fine-tune on target task (e.g., 20,000 steps)

# Evaluate the fine-tuned model
mean_reward, std_reward = evaluate_policy(model, target_env, n_eval_episodes=10)
print(f"Mean reward in target environment ({target_env_name}): {mean_reward} +/- {std_reward}")

# Close environments
source_env.close()
target_env.close()
```
